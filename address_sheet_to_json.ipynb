```python
import geocoder
import json
import xlrd
import re
import csv

```

```python
wb = xlrd.open_workbook("Address-datacleaning.xls")
```

```python
first_sheet = wb.sheet_by_index(0)
first_sheet.name
```

```python
empty_latlng_address = []
address_keys = []
address_values=[]
i = 1
for row_index in range(3001,4501):
    cell_value = first_sheet.cell(row_index,1)
    cell_value = cell_value.value.encode('ascii','ignore')
    g1 = geocoder.google(cell_value)
    
    j = 1
    
    if ";" in cell_value:
        for address_key in cell_value.split(";"):
            g = geocoder.google(address_key)
            print(i,j,address_key,g.latlng)
            print("*****************************\n")
            if g.latlng:
                address_values.append(g.latlng)
                address_keys.append(address_key)
            else:
                empty_latlng_address.append(address_key)
            j+=1
            
    else:
        g = geocoder.google(cell_value)
        print(i,cell_value,g1.latlng)
        print("****************************\n")
        if g.latlng:
            address_keys.append(cell_value)
            address_values.append(g.latlng)
        else:
            empty_latlng_address.append(cell_value)
    i+=1
```

```python
len(empty_latlng_address)# address without latlong from excel sheet (List 1 -Address)
```

```python
len(address_values) # address with latlong from excel sheet (List 1 -Address)
```

```python
address_dict = dict(zip(address_keys,address_values))

address_dict
```

```python
with open('empty_latlng_address.csv','wb') as data:
    writer = csv.writer(data,lineterminator='\n')
    for val in empty_latlng_address:
        writer.writerow([val])
    
```

```python
import json
with open('address.json','w') as v:
    json.dump(address_dict,v)
```
