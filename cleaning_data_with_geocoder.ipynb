```python
import geocoder
import json
import xlrd
import csv
import xlwt
from xlutils import copy
```

```python
# load organization json file and return
def read_clean_org_data(org_json_file_path):
    with open(org_json_file_path) as data:
        return json.load(data)
# load address json file and return
def read_address_data(addr_json_file_path):
    with open(addr_json_file_path) as data:
        return json.load(data)

# update existing excel sheet by adding sorted data into desired sheet
def update_excel_sheet(excel_file,address_data,org_data):
    rb = xlrd.open_workbook(excel_file)
    wb = copy.copy(rb)
    sheet= wb.get_sheet(2)
    orgs =[]
    addrs = []
   
    for org_key,org_value in org_data.items():
        for address_key,address_value in address_data.items():
            if org_data[org_key] == address_data[address_key]:
                orgs.append(org_key)
                addrs.append(address_key)
                print("[>>>>]: {}\n{}\n".format(org_key,org_value))
                print("[>>>>]: {}\n{}\n".format(address_key,address_value))
                print("****************************************************************\n")
                
    for x in range(len(orgs)):
        sheet.write(x+1,0,orgs[x])
        sheet.write(x+1,1, addrs[x])
    wb.save(excel_file)

def main():
    excel_file = 'Address-datacleaning.xls'
    org_file = 'org.json'
    address_file = 'address.json'
    org_data = read_clean_org_data(org_file)
    address_data = read_address_data(address_file)
    update_excel_sheet(excel_file,address_data,org_data)
    
if __name__ == "__main__":
    main()
    
```
