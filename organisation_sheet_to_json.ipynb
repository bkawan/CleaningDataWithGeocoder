```python
import geocoder
import json
import xlrd
```

```python
wb = xlrd.open_workbook("Address-datacleaning.xls")
second_sheet = wb.sheet_by_index(1)
```

```python
second_sheet.name
```

```python
org_keys = []
org_values = []
```

```python
for row_index in range(second_sheet.nrows):
    value = second_sheet.cell(row_index,1)
    g = geocoder.google(value.value)
    org_keys.append(str(value.value))
    org_values.append(g.latlng)
    print(str(value.value),g.latlng)
    
```

```python
org_dict = dict(zip(org_keys,org_values))
```

```python
org_dict
```

```python
with open ("org.json",'w') as v:
    json.dump(org_dict,v)
```
